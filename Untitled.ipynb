```python
from itertools import product
 
def cartesian_product(arr1, arr2, arr3):
 
    # return the list of all the computed tuple
    # using the product() method
    return list(product(arr1, arr2, arr3))
```

```python
numpy_array = cartesian_product(arr1, arr2, arr3)
```

```python
mobile = application.FLAG_MOBIL.unique() 
```

```python
work_phone = application.FLAG_WORK_PHONE.unique() 
```

```python
phone = application.FLAG_PHONE.unique() 
```

```python
email = application.FLAG_EMAIL.unique() 
```

```python
# Function to define cartesian product
def cartesian_product_communication(arr1, arr2, arr3, arr4):
 
    # return the list of all the computed tuple
    # using the product() method
    return list(product(arr1, arr2, arr3, arr4))
```

```python
# Calling function to get all posible collection in array format
cartesian_product_communication(mobile, work_phone, phone, email)
```

```python
# stoing result above in varaible
numpy_array_communication = cartesian_product_communication(mobile, work_phone, phone, email)
```

```python
# Making dataframe out of array
communication = pd.DataFrame(numpy_array_communication, columns = ['has_mobile', 'has_work_phone', 'has_phone', 'has_email'])
```

```python
# calling dataframe
communication
```

```python
# converting dataframe to csv
communication.to_csv('communication.csv')
```
